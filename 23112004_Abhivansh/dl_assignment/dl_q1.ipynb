```python
text_data = " ".join(training_df['text'].tolist())
chars = sorted(list(set(text_data)))
char_to_idx={ch: idx for idx, ch in enumerate(chars)}
idx_to_char={idx: ch for idx, ch in enumerate(chars)}
```

```python
class modelQ1(nn.Module):
  def __init__(self,input_size,hidden_size,output_size):
    super(modelQ1, self).__init__()
    self.hidden_size=hidden_size
    self.rnn1=nn.RNN(input_size,hidden_size,nonlinearity="tanh",num_layers=2)
    self.rnn2=nn.RNN(hidden_size,hidden_size,nonlinearity="tanh",num_layers=2)
    self.fc=nn.Linear(hidden_size,output_size)

  def forward(self,x,hidden):
    batch_size=x.size(0)
    out,hidden=self.rnn1(x,hidden)
    out,hidden=self.rnn2(out,hidden)
    out=self.fc(out)
    return out,hidden

  def generate(self,prompt,context_window,length,char_to_idx,idx_to_char,device):
        self.eval()
        with torch.no_grad():
            input=torch.zeros(1,1,len(char_to_idx)).to(device)
            input[0,0,char_to_idx[prompt]]=1
            hidden=torch.zeros(2,1,self.hidden_size).to(device)
            output=prompt
            for _ in range(length):
                out, hidden=self.rnn1(input,hidden)
                out, hidden=self.rnn2(out, hidden)
                out=self.fc(out)
                out=torch.softmax(out, dim=2)
                char_idx=torch.multinomial(out[0, -1], 1).item()
                output += idx_to_char[char_idx]
                input=torch.zeros(1,1,len(char_to_idx)).to(device)
                input[0, 0,char_idx]=1
            return output
```

```python
def get_batches_from_dataframe(training_df,batch_size,seq_length,char_to_idx, device):
    text_data=training_df['text'].tolist()
    num_rows=len(text_data)

    while True:
        x_batch=[]
        y_batch=[]
        for _ in range(batch_size):
            row_idx=np.random.randint(0, num_rows)
            text_row = text_data[row_idx]
            if len(text_row) <= seq_length + 1:
                continue #leaving the short sequences


            start_indx = np.random.randint(0, len(text_row) - seq_length - 1)
            seq_x=text_row[start_indx:start_indx + seq_length]
            seq_y= text_row[start_indx + 1:start_indx + seq_length + 1]
            x_batch.append([char_to_idx[ch] for ch in seq_x])
            y_batch.append([char_to_idx[ch] for ch in seq_y])

        x_batch = torch.tensor(x_batch).to(device)
        y_batch = torch.tensor(y_batch).to(device)

        x_onehot = torch.zeros(x_batch.size(0), x_batch.size(1), len(char_to_idx)).to(device)
        for i in range(x_batch.size(0)):
            for j in range(x_batch.size(1)):
                x_onehot[i, j, x_batch[i, j]] = 1

        yield x_onehot, y_batch
```

```python
input_size=len(chars)
hidden_size=256
output_size=len(chars)
learning_rate=0.002
num_epochs=1
batch_size=10
seq_length=10

device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
model=modelQ1(input_size, hidden_size, output_size).to(device)

criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    model.train()
    for x, y in get_batches_from_dataframe(training_df, batch_size, seq_length, char_to_idx, device):
        hidden = torch.zeros(2, x.size(0), hidden_size).to(device)
        print(hidden.size())
        x=x.to(device)
        t=hidden.size()
        optimizer.zero_grad()
        output,_= model(x,hidden)
        loss = criterion(output.view(-1, len(char_to_idx)), y.view(-1))
        loss.backward()
        optimizer.step()

    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
print("Training complete")
```
